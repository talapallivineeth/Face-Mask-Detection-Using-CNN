Importing Modulus

```python
#importing modules
from tensorflow.keras.models import model_from_json
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Loading the Model

```python
json_file = open('model.json', 'r')
loaded_model_json = json_file.read() #reading the json file
json_file.close()
loaded_model = model_from_json(loaded_model_json)   
loaded_model.load_weights("model.h5")  # load weights into new model
```

Predicting the model on sample images

```python
face_clsfr=cv2.CascadeClassifier('face_xml.xml') #Loading the face Detector classifier
labels_dict={0:'MASK',1:'NO MASK'}     #Labelling the categories  
color_dict={0:(0,255,0),1:(0,0,255)}   #Defining colors for categories
```

```python
img=cv2.imread('C:/Users/vinu/Desktop/train/with mask/masked000000.jpg') #reading the image
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)   #converting into the gray scale
faces=face_clsfr.detectMultiScale(gray_img,1.05,5)   #detecting the face coordinates
for (x,y,w,h) in faces:    
        face_img=img[y:y+w,x:x+w]
        resized=cv2.resize(face_img,(224,224))  #re-sizing the image
        new_img=preprocess_input(resized)   
        reshaped=np.reshape(new_img,(1,224,224,3)) #re-shape the numpy array of image pixels 
        result=loaded_model.predict(reshaped)  #predicting the category of the image 

        label=np.argmax(result,axis=1)[0]
      
        cv2.rectangle(img,(x,y),(x+w,y+h),color_dict[label],2)  #drawing the countours around the face
        cv2.rectangle(img,(x,y-40),(x+w,y),color_dict[label],-1)
        cv2.putText(img, labels_dict[label], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
plt.imshow(img)
plt.title('my picture')
plt.show()
```
